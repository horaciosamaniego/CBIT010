### Ejercicio 7: Índice de Shannon ($H'$) con bucle

En la Semana 3 calcularon $H'$ a mano para 3 especies. Ahora háganlo
para **cualquier número de especies** usando un bucle.

$$H' = −\sum_{i=1}^{S} p_i \times ln(p_i)$$

donde $p_i =$ abundancia de la especie $i$ / total de individuos

```python
import math

abundancias = [45, 28, 67, 12, 33, 8, 52]
nombres = ["Coigüe", "Canelo", "Luma", "Mañío", "Tepa", "Tineo", "Ulmo"]

# TU CÓDIGO AQUÍ
total = sum(abundancias)    # aquí sí usamos sum() — ya saben cómo funciona

H = 0
for i in range(len(abundancias)):
    if abundancias[i] > 0:     # ln(0) no existe
        p = abundancias[i] / total
        H -= ___               # p * math.log(p)

H_bits = H / math.log(2)      # convertir nats a bits

print(f"{'Especie':<12} {'N':>5} {'pi':>8}")
print("-" * 28)
for i in range(len(nombres)):
    p = abundancias[i] / total
    print(f"{nombres[i]:<12} {abundancias[i]:>5} {p:>8.3f}")

print(f"\n{'Total':<12} {total:>5}")
print(f"\nH' = {H:.4f} nats")
print(f"H' = {H_bits:.4f} bits")
print(f"Riqueza S = {len(abundancias)}")
print(f"H' máximo = ln(S) = {math.log(len(abundancias)):.4f} nats")
print(f"Equitatividad J = H'/H'max = {H / math.log(len(abundancias)):.4f}")
```
