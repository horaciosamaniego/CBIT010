## 🏆 Ejercicio 6 (Desafío): Diversidad por parcela

Calcula el índice de Shannon H' para **cada parcela** por separado.
¿Cuál parcela tiene mayor diversidad?

```python
import math

# Paso 1: construir un diccionario de conteos por parcela
# Estructura: {1: {"N. dombeyi": 1, "D. winteri": 1, ...}, 2: {...}, 3: {...}}

conteos_por_parcela = {}

for a in arboles:
    p = a["parcela"]
    sp = a["especie"]

    if p not in conteos_por_parcela:
        conteos_por_parcela[p] = {}

    conteos_por_parcela[p][sp] = conteos_por_parcela[p].get(sp, 0) + 1

# Paso 2: calcular H' para cada parcela
# TU CÓDIGO AQUÍ

print("Diversidad por parcela:\n")
for parcela in sorted(conteos_por_parcela.keys()):
    conteos = conteos_por_parcela[parcela]
    total = sum(conteos.values())
    S = len(conteos)

    H = 0
    for sp, n in conteos.items():
        pi = n / total
        H -= pi * math.log(pi)

    H_max = math.log(S) if S > 1 else 0
    J = H / H_max if H_max > 0 else 0

    print(f"  Parcela {parcela}: S={S}, N={total}, H'={H:.3f}, J={J:.3f}")
    for sp, n in conteos.items():
        print(f"    {sp}: {n}")
    print()
```
