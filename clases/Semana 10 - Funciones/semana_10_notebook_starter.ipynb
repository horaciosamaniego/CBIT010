## Ejercicio 4: Funciones de diversidad 🌿

El objetivo: escribir cada cálculo **una vez** como función, y luego
combinarlas en un reporte.

```
import math

# 4a. Shannon H'
def shannon_entropy(abundancias):
    """
    Índice de Shannon (H') en nats.
    
    Parámetros:
        abundancias: list[int] — abundancia de cada especie
    Retorna:
        float — H' (logaritmo natural)
    """
    total = sum(abundancias)
    H = 0
    for n in abundancias:
        if n > 0:
            p = n / total
            H -= ___   # p * math.log(p)
    return H

# Probar
print(f"[45,28,67,15,33] → H' = {shannon_entropy([45,28,67,15,33]):.4f}")
print(f"[10,10,10,10,10] → H' = {shannon_entropy([10,10,10,10,10]):.4f}")  # máximo
print(f"[100,1,1,1,1]    → H' = {shannon_entropy([100,1,1,1,1]):.4f}")     # bajo
```

```
# 4b. Equitatividad
def equitatividad(abundancias):
    """J = H' / H'_max. Retorna 0 si hay ≤1 especie."""
    S = len(abundancias)
    if S <= 1:
        return 0.0
    H = ___   # llamar a shannon_entropy()
    H_max = math.log(S)
    return H / H_max

# Probar
print(f"[10,10,10,10,10] → J = {equitatividad([10,10,10,10,10]):.4f}")  # 1.0
print(f"[100,1,1,1,1]    → J = {equitatividad([100,1,1,1,1]):.4f}")     # bajo
```

```
# 4c. Simpson (1-D)
def simpson_index(abundancias):
    """Índice de Simpson (1 - D). Rango: 0 a 1."""
    total = sum(abundancias)
    D = sum((n / total) ** 2 for n in abundancias)
    return 1 - D

# Probar
print(f"[10,10,10,10,10] → 1-D = {simpson_index([10,10,10,10,10]):.4f}")  # 0.8
print(f"[100,1,1,1,1]    → 1-D = {simpson_index([100,1,1,1,1]):.4f}")
```

```
# 4d. Reporte integrado — composición de funciones
def reporte_diversidad(nombre, abundancias):
    """Imprime un reporte completo de diversidad."""
    S = len(abundancias)
    N = sum(abundancias)
    H = shannon_entropy(abundancias)
    J = equitatividad(abundancias)
    D = simpson_index(abundancias)
    
    print(f"{'='*44}")
    print(f"  Reporte de diversidad: {nombre}")
    print(f"{'='*44}")
    print(f"  Riqueza (S):       {S}")
    print(f"  Abundancia (N):    {N}")
    print(f"  Shannon (H'):      {H:.4f} nats")
    print(f"  Equitatividad (J): {J:.4f}")
    print(f"  Simpson (1-D):     {D:.4f}")
    print(f"{'='*44}\n")

# Probar
reporte_diversidad("Parcela equitativa", [10, 10, 10, 10, 10])
reporte_diversidad("Parcela dominada",   [90, 3, 2, 3, 2])
reporte_diversidad("Alerce Andino",      [45, 28, 67, 15, 33])
```
