## PARTE A — Demostración guiada

```
import pandas as pd
import matplotlib.pyplot as plt
import math
from google.colab import drive

drive.mount('/content/drive')

ruta = "/content/drive/MyDrive/CBIT010_Archivos_ayudantia/Semana13/monitoreo_biodiversidad_2025.csv"
df = pd.read_csv(ruta)

# Función de diversidad de la Semana 10
def shannon_entropy(abundancias):
    """Índice de Shannon H' en nats."""
    total = sum(abundancias)
    H = 0
    for n in abundancias:
        if n > 0:
            p = n / total
            H -= p * math.log(p)
    return H
```

### Paso 1 · Cargar y mirar
Siempre mirar los datos antes de analizarlos.

```
print(f"Dimensiones: {df.shape}")
print(f"\nColumnas: {list(df.columns)}")
print(f"\nTipos:\n{df.dtypes}")
df.head(10)
```

### Paso 2 · Diagnóstico de calidad
¿Qué problemas tienen estos datos?

```
print("Valores faltantes por columna:")
print(df.isna().sum())
print(f"\nFilas duplicadas: {df.duplicated().sum()}")
print(f"\nEstadísticas de abundancia:")
print(df["abundancia"].describe())
print(f"\nEspecies registradas:")
print(sorted(df["especie"].unique()))
```

**Problemas detectados:**
1. Valores faltantes (NaN) en `abundancia`
2. Una fila duplicada
3. Un valor atípico: 999 (claramente un error de tipeo)
4. Una especie escrita en minúscula ("chucao" en vez de "Chucao")

### Paso 3 · Limpieza
Cada decisión se justifica.

```
# Eliminar duplicados
df = df.drop_duplicates()

# Convertir abundancia a número (vacíos → NaN)
df["abundancia"] = pd.to_numeric(df["abundancia"], errors="coerce")

# El 999 es un error de tipeo → lo tratamos como dato faltante
df.loc[df["abundancia"] > 100, "abundancia"] = None

# Rellenar NaN con la mediana (más robusta que la media ante outliers)
df["abundancia"] = df["abundancia"].fillna(df["abundancia"].median())

# Corregir inconsistencia de mayúsculas
df["especie"] = df["especie"].str.capitalize()

# Verificar que quedó limpio
print("Después de limpiar:")
print(f"  Filas: {len(df)}")
print(f"  NaN: {df.isna().sum().sum()}")
print(f"  Abundancia máx: {df['abundancia'].max()}")
print(f"  Especies: {df['especie'].nunique()}")
```

### Paso 4 · Análisis

```
print("Riqueza de especies por sitio:")
print(df.groupby("sitio")["especie"].nunique())

print("\nAbundancia total por grupo taxonómico:")
print(df.groupby("grupo")["abundancia"].sum())

print("\nDiversidad de Shannon por sitio:")
for sitio, grupo in df.groupby("sitio"):
    abund = grupo.groupby("especie")["abundancia"].sum()
    H = shannon_entropy(abund.values)
    print(f"  {sitio}: S={len(abund)}, H'={H:.3f}")
```

### Paso 5 · Visualización

```
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

riqueza = df.groupby("sitio")["especie"].nunique().sort_values()
riqueza.plot(kind="barh", ax=axes[0], color="#2E7D32", edgecolor="white")
axes[0].set_title("Riqueza de especies por sitio")
axes[0].set_xlabel("Nº de especies")

df.groupby("grupo")["abundancia"].sum().plot(
    kind="bar", ax=axes[1], color="#1565C0", edgecolor="white")
axes[1].set_title("Abundancia por grupo taxonómico")
axes[1].set_ylabel("Individuos")
axes[1].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()
```

### Paso 6 · Interpretación

*Ejemplo:* El Parque Alerce Costero es el sitio con mayor riqueza y mayor
diversidad de Shannon. Las aves dominan en abundancia. Limitaciones: el
número de visitas por sitio no es uniforme, y la detección de anfibios
probablemente está subestimada (sesgo de detección, Sem. 3).
