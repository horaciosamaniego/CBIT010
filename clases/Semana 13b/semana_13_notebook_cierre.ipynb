```
import pandas as pd
import matplotlib.pyplot as plt
import math
from google.colab import drive

drive.mount('/content/drive')

# Ajusta la ruta a tu carpeta
ruta = "/content/drive/MyDrive/CBIT010_Archivos_ayudantia/Semana13/monitoreo_biodiversidad_2025.csv"
df = pd.read_csv(ruta)

# Función de diversidad (de la Semana 10 — pueden copiarla o reescribirla)
def shannon_entropy(abundancias):
    """Índice de Shannon H' en nats."""
    total = sum(abundancias)
    H = 0
    for n in abundancias:
        if n > 0:
            p = n / total
            H -= p * math.log(p)
    return H
```
